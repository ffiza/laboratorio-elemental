# Histogramas

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Configurar los gr√°ficos
mpl.rcParams['font.size'] = 8
mpl.rcParams['figure.dpi'] = 300
mpl.rcParams['figure.figsize'] = [3, 2.5]
```

```python
# Cargar datos
data = np.loadtxt('data.csv')
```

## Histograma con cuentas en cada bin

```python
fig, ax = plt.subplots()
ax.tick_params(which='both', direction="in")
ax.set_axisbelow(True)
ax.grid(True, linestyle='--', lw=.5)
ax.minorticks_on()
ax.set_xlabel('Tiempo [s]')
ax.set_ylabel('Occurencias')
ax.set_xlim(.8, 1.5)
ax.set_xticks([.9, 1, 1.1, 1.2, 1.3, 1.4, 1.5])
ax.set_ylim(0, 80)

plt.hist(data, color='tab:red', alpha=0.5)

plt.show()
```

## Densidad de probabilidad

```python
_, ax = plt.subplots()
ax.tick_params(which='both', direction="in")
ax.set_axisbelow(True)
ax.grid(True, linestyle='--', lw=.5)
ax.minorticks_on()
ax.set_xlabel('Tiempo [s]')
ax.set_ylabel('Densidad de probabilidad')
ax.set_xlim(.8, 1.5)
ax.set_xticks([.9, 1, 1.1, 1.2, 1.3, 1.4, 1.5])
ax.set_ylim(0, 4)

plt.hist(data, color='tab:red', density=True, alpha=0.5)

plt.show()
```
