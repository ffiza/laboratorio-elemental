# Histogramas

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Configurar los gr√°ficos
params = {'figure.dpi': 300,
          'figure.figsize': [3, 2.5],
          'font.size': 8,
          'xtick.top': 'on',
          'xtick.minor.bottom': 'on',
          'xtick.minor.top': 'on',
          'xtick.direction': 'in',
          'ytick.right': 'on',
          'ytick.minor.left': 'on',
          'ytick.minor.right': 'on',
          'ytick.direction': 'in',
          'savefig.dpi': 300,
          'savefig.bbox': 'tight',
          'savefig.pad_inches': .02,
          'axes.axisbelow': True,
          'axes.grid': True,
          'grid.linestyle': '--',
          'grid.linewidth': .5}
plt.rcParams.update(params)
```

```python
# Cargar datos
data = np.loadtxt('data.csv')
```

## Histograma con cuentas en cada bin

```python
fig, ax = plt.subplots()
ax.minorticks_on()
ax.set_xlabel('Tiempo [s]')
ax.set_ylabel('Occurencias')
ax.set_xlim(.8, 1.5)
ax.set_xticks([.9, 1, 1.1, 1.2, 1.3, 1.4, 1.5])
ax.set_ylim(0, 80)

plt.hist(data, color='tab:red', alpha=0.7)

plt.show()
```

## Densidad de probabilidad

```python
_, ax = plt.subplots()
ax.minorticks_on()
ax.set_xlabel('Tiempo [s]')
ax.set_ylabel('Densidad de probabilidad')
ax.set_xlim(.8, 1.5)
ax.set_xticks([.9, 1, 1.1, 1.2, 1.3, 1.4, 1.5])
ax.set_ylim(0, 4)

plt.hist(data, color='tab:red', density=True, alpha=0.7)

plt.show()
```
